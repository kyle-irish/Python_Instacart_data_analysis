```python
#import merged orders and products data with added price range and busy time variables as ords_prods_merge
ords_prods_merge = pd.read_pickle(os.path.join(path, 'prepared data' , 'order product merge with price ranges and busy counts.pkl'))
```

```python
#create df, a subset of the first million rows of ords_prods_merge
df = ords_prods_merge[:1000000]
```

```python
#running groupby() by itself to create an object
df.groupby('product_name')
```

```python
#running grouby() with agg() to find the mean order_number across departments
df.groupby('department_id').agg({'order_number':['mean']})
```

```python
#get the min, max, and mean of 'order_number' by department
df.groupby('department_id').agg({'order_number':['mean','min', 'max']})
```
