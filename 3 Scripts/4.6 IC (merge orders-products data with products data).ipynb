```python
#remove the merge column from df_ords_prods
df_ords_prods = df_ords_prods.drop(columns =['_merge'])
```

```python
#df_prods and df_ords_prods both have a product_id column. will merge on that with outer join and merge flag count 
df_ords_prods_total = df_ords_prods.merge(df_prods, on ='product_id', how = 'outer', indicator = True)
```
